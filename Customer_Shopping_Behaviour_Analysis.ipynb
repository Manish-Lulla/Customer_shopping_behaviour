```python
## Read the dataset
import pandas as pd
import sqlite3

df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
## Display basic information about the dataset
df.info()
```

```python
## Display statistical summary of the dataset
df.describe(include='all')
```

```python
## Check for missing values
df.isnull().sum()
```

```python
## We will replace missing values in 'Review Rating' with the median rating of that product category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
## Rename columns for better readability and consistency
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.head()
```

```python
## Create a column age_group based on age
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
df[['age', 'age_group']].head(10)
```

```python
## Create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
## Display first 10 rows of discount_applied and promo_code_used columns
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
## Compare two columns discount_applied and promo_code_used
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
## Drop promo_code_used column as both columns are identical
df = df.drop(columns=['promo_code_used'])
df.columns
```

```python
# Create SQLite database in memory

conn = sqlite3.connect(':memory:')
```

```python
# Convert CSV to SQL table

df.to_sql('shopping_behavior', conn, index=False, if_exists='replace')
```

```python
# Q1 Total revenue by gender

query = """
SELECT
    "Gender",
    SUM("Purchase Amount (USD)") AS revenue
FROM shopping_behavior
GROUP BY "Gender";
"""

pd.read_sql_query(query, conn)
```

```python
# Q2 Customers who used discount and spent above average

query = """
SELECT
    "Customer ID",
    "Purchase Amount (USD)"
FROM shopping_behavior
WHERE "Discount Applied" = 'Yes'
AND "Purchase Amount (USD)" >= (
    SELECT AVG("Purchase Amount (USD)")
    FROM shopping_behavior
);
"""

pd.read_sql_query(query, conn)
```

```python
# Q3 Top 5 products with highest review rating

query = """
SELECT
    "Item Purchased",
    ROUND(AVG("Review Rating"),2) AS avg_rating
FROM shopping_behavior
GROUP BY "Item Purchased"
ORDER BY avg_rating DESC
LIMIT 5;
"""

pd.read_sql_query(query, conn)
```

```python
# Q4 Average purchase amount by shipping type

query = """
SELECT
    "Shipping Type",
    ROUND(AVG("Purchase Amount (USD)"),2) AS avg_purchase
FROM shopping_behavior
WHERE "Shipping Type" IN ('Standard','Express')
GROUP BY "Shipping Type";
"""

pd.read_sql_query(query, conn)
```

```python
# Q5 Subscribers vs Non-Subscribers spending


query = """
SELECT
    "Subscription Status",
    COUNT("Customer ID") AS total_customers,
    ROUND(AVG("Purchase Amount (USD)"),2) AS avg_spend,
    ROUND(SUM("Purchase Amount (USD)"),2) AS total_revenue
FROM shopping_behavior
GROUP BY "Subscription Status"
ORDER BY total_revenue DESC;
"""

pd.read_sql_query(query, conn)
```

```python
# Q6 Products with highest discount rate


query = """
SELECT
    "Item Purchased",
    ROUND(
        100.0 * SUM(CASE WHEN "Discount Applied"='Yes' THEN 1 ELSE 0 END)/COUNT(*),
        2
    ) AS discount_rate
FROM shopping_behavior
GROUP BY "Item Purchased"
ORDER BY discount_rate DESC
LIMIT 5;
"""

pd.read_sql_query(query, conn)
```

```python
# Q7 Customer segmentation

query = """
WITH customer_type AS (
    SELECT
        "Customer ID",
        "Previous Purchases",
        CASE
            WHEN "Previous Purchases" = 1 THEN 'New'
            WHEN "Previous Purchases" BETWEEN 2 AND 10 THEN 'Returning'
            ELSE 'Loyal'
        END AS customer_segment
    FROM shopping_behavior
)
SELECT
    customer_segment,
    COUNT(*) AS num_customers
FROM customer_type
GROUP BY customer_segment;
"""

pd.read_sql_query(query, conn)
```

```python
# Q8 Top 3 products in each category


query = """
WITH item_counts AS (
    SELECT
        "Category",
        "Item Purchased",
        COUNT("Customer ID") AS total_orders,
        ROW_NUMBER() OVER (
            PARTITION BY "Category"
            ORDER BY COUNT("Customer ID") DESC
        ) AS item_rank
    FROM shopping_behavior
    GROUP BY "Category","Item Purchased"
)
SELECT
    item_rank,
    "Category",
    "Item Purchased",
    total_orders
FROM item_counts
WHERE item_rank <= 3;
"""

pd.read_sql_query(query, conn)
```
